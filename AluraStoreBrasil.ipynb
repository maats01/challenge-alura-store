### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

# Análise exploratória dos dados

```python
loja.info()
```

```python
loja2.info()
```

```python
loja3.info()
```

```python
loja4.info()
```

Como os quatro dataframes possuem a mesma estrutura, podemos concatená-los para facilitar a manipulação deles.

```python
# adicionando uma coluna identificadora
loja['Loja'] = 'Loja 1'
loja2['Loja'] = 'Loja 2'
loja3['Loja'] = 'Loja 3'
loja4['Loja'] = 'Loja 4'

df = pd.concat([loja, loja2, loja3, loja4])
df.head(10)
```

```python
df.info()
```

Como podemos ver, o dataframe da união das lojas possui **9435** instâncias e todas elas são de valores não-nulos, então não precisamos nos preocupar com esse tratamento.

Para manter a consistência dos valores numéricos, podemos arredondar os valores da coluna **Frete** para duas casas decimais, ficando consistente com a coluna **Preço**.

```python
df['Frete'] = df['Frete'].round(2)
df.head()
```

#1. Análise do faturamento



```python
faturamentos = df.groupby('Loja')['Preço'].sum().reset_index(name='Preço')

display(faturamentos.style.format(
    {'Preço': '{:,.2f}'},
    thousands='.',
    decimal=','
))
```

```python
ax = faturamentos.plot.bar(color='teal', figsize=(10,6))

ax.bar_label(ax.containers[0], fmt=lambda x: f'R$ {(x/pow(10,6)).round(2)}M')
ax.set_xticklabels(faturamentos['Loja'])
plt.title('Faturamentos de cada loja')
plt.xlabel('Lojas')
plt.ylabel('Faturamento Total (R$)')

plt.show()
```

# 2. Vendas por Categoria

```python
df.groupby('Loja')['Categoria do Produto'].value_counts()
```

Todas as lojas possuem as mesmas três categorias mais vendidas, **Moveis, Eletrônicos e Eletrodomésticos**, indicando que grande parte do faturamento provém de vendas de produtos dessas categorias. Para confirmar isso, podemos analisar o faturamento relativo a cada categoria.

```python
faturamento_por_categoria = df.groupby('Categoria do Produto')['Preço'].sum().reset_index(name='Faturamento Total').sort_values(by='Faturamento Total', ascending=False)
display(faturamento_por_categoria.style.format(
    {'Faturamento Total': '{:,.2f}'},
    thousands='.',
    decimal=','
))
```

```python
faturamento_por_categoria.set_index('Categoria do Produto').plot(
    y='Faturamento Total',
    kind='pie',
    figsize=(12,12),
    autopct='%1.1f%%',
    legend=True,
    title='Faturamento por Categoria',
    ylabel='',
    labeldistance=None
)
```

Analisando o gráfico acima, fica fácil notar que os produtos das categorias **Eletronicos, Eletrodomesticos e Moveis** possuem um impacto significativo de **85%** no faturamento total da AluraStore. Comprovando a importância dessas três categorias para o desempenho das lojas.

# 3. Média de Avaliação das Lojas


Outra métrica importante a ser analisada é a avaliação média dos clientes, pois com ela podemos inferir a satisfação do cliente com a sua compra.

```python
df.groupby('Loja')['Avaliação da compra'].mean().round(2).sort_values(ascending=False)
```

Todas as lojas possuem avaliações médias muito parecidas, destacando-se a **Loja 3** com a maior avaliação média de **4,05** e a **Loja 1**, que apesar de possuir o maior faturamento das quatro lojas, possui a menor avaliação média de **3,98**.

# 4. Produtos Mais e Menos Vendidos

```python
vendas_por_categoria = df.groupby('Categoria do Produto').size().reset_index(name='Total de vendas').sort_values(by='Total de vendas', ascending=False)
vendas_por_categoria
```

As unidades vendidas que pertencem as categorias **Móveis**, **Eletrodomesticos** e **Eletrônicos**, correspondem a aproximadamente **51%** de todas as unidades vendidas da AluraStore.

Praticamente, metade das unidades vendidas correspondem a **85%** do faturamento total da AluraStore.

# 5. Frete Médio por Loja

```python
frete_medio = df.groupby('Loja')['Frete'].mean().reset_index(name='Frete médio')
display(frete_medio.style.format(
    {'Frete médio': '{:,.2f}'},
    thousands='.',
    decimal=','
))
```

```python
fig, axes = plt.subplots()

axes.stem(frete_medio['Loja'], frete_medio['Frete médio'], basefmt=' ')

axes.set_ylim(0)
plt.title('Fréte médio por loja')
plt.xlabel('Loja')
plt.ylabel('Valor médio de frete')
plt.xticks(frete_medio['Loja'])
```

Os valores médios de frete das lojas são muito parecidos, destacando-se somente a **Loja 4**, com o menor frete médio de **R$ 31,28**.

# Relatório Final

Análisando diferentes aspectos da AluraStore, a loja com a menor eficiência e a que deve ser vendida é a **Loja 4**. A justificativa se dará com uma breve análise de diferentes perspectivas da AluraStore:
- **Faturamento:** a Loja 4 possui o menor faturamento de todas as quatro lojas, faturando cerca de R\$ 150.000,00 a menos que a Loja 1. Isso já é um grande indicador da sua baixa performance.

Isso é um indicador da sua baixa performance no faturamento total.
- **Avaliação média dos clientes:** as quatro lojas possuem avaliações médias extremamente parecidas, o que nos indica que a má performance da Loja 4 não resulta da satistfação dos clientes (inclusive, a Loja 1 que possui o maior faturamento, possui uma avaliação média menor que a da Loja 4).
- **Frete médio:** um ponto positivo da Loja 4 é que ela se destaca em relação as outras lojas por possuir o menor frete médio (R\$ 31,28).
